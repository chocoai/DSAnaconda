```python
import requests
from bs4 import BeautifulSoup
res = requests.get('http://news.qq.com/')
soup = BeautifulSoup(res.text,'html.parser')

newsarry = []
for news in soup.select('.Q-tpList .text'):
    newsarry.append({'title': news.select('a')[0].text,'url':news.select('a')[0]['href']})
```

```python
import pandas
newsdf = pandas.DataFrame(newsarry)
newsdf
```

```python
newsdf.to_excel('news.xlsx')
```
